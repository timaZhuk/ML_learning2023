* Associated with each photon is a wave described by complex wavefunction $AU(r)exp(i\omega t)$ (general view). But when a photon impinges on a detector of small area $dA$ located **normal** to the direction of propagation at the position **r**, its individualitity causes it to be either wholly detected or not detected at all. The location at which the photon is registered is not precisely determined. The **probability per unit volume** of finding a photon in a given region of space at an instant of time is proportional to **the number of photons per unit volume** at that time:
$$$$
$$ \frac{Probability}{V} \infty \frac{N}{V}$$
$$$$
* The **number of photons per unit volume** is proportional to the **intensity** of the radiation:
$$$$
$$\frac{N}{V} \infty I$$
$$$$
* Now, let’s form a connection between the particle model and the wave model by recalling that the **intensity of electromagnetic radiation** is proportional to the **square of the electric field amplitude E** for the electromagnetic wave:
$$$$
$$ I \infty E^2 $$

* Equating the beginning and the end of this series of proportionalities gives:
$$$$
$$ \frac{Probability}{V} \infty E^2$$

* Therefore, for electromagnetic radiation, the probability per unit volume of finding a particle associated with this radiation (the photon) is proportional to the square of the amplitude of the associated electromagnetic wave. The amplitude of the **deBroglie wave** associated with a particle is *not a measurable quantity* because the wave function representing a particle is generally a **complex function** $AU(r)exp(i\omega t)$. In contrast, the electric field for an electromagnetic wave is a real function. In physics are many models have as mathematical represtationa as a real measurable quantities.

## absolute refractive index
* The absolute refractive index characterizes the electrical and magnetic properties of the medium, it depends on the wavelength of light:
$$$$
$$n_{\lambda} = \sqrt{\epsilon \mu}$$
$$$$
In the optical region of the spectrum for all substances $\mu \cong 1$, therefore $n_{\lambda} \cong \sqrt{\epsilon}$.
* $ \epsilon $ - is determined by the interaction of the electromagnetic field of light wave with atoms and molecules of matter. The propagation of light in medium is related to the polarizability of the molecules in substance. Under the influence of a field, the elctrons of atoms or molecules make a forced oscillations. When the oscillation frequency of the field coincides with the natural oscillation frequency of the electrones, resonance and absorbtion of light occure. At frequencies close to resonant, a strong depends of  $ \epsilon $ on $\lambda$ appears.
## Dependence of index of refraction from density of substance
* As the density increases, the refractive index (n) increases. The dependence is different for different aggregate states of substance. So for liquids, the Gladson-Dal formula is used: 
$$$$
$$ n-1 = r\rho$$
$$$$
$\rho$ - density of substance, r - specific refraction or refractive power of substance. 
$$r = (n-1)\frac{1}{\rho}$$
* In experimental practice the Lorentz law is used, which is ideally for isotropic media (gases, non polar liquids) in the visible and ultraviolet ranges of electromagnetic radiaation:
$$$$
$$ r = \frac{n^2-1}{n^2+2}\frac{1}{\rho}$$
$$$$
Refraction is a measure of the electronic polarizability of a substance, characterizing its physical state.

## For multiple layers  
If we have multiple layers with the different thickness for different materials. We can modify the formula:
$$ \\   N_i = N_0\frac{\Delta z_i}{nh} \frac{r+x_i}{r} e^{-(\mu/\rho)_i \rho x_i}   $$
This gives us some restrictions for the radius of active exposure and thickness of multiple layers $r > x_i$ and $r≤ \sum_{i}^{n}x_i$
************************************************
$$ \\   N_i = N_0\frac{\Delta z_i}{nh} (1 + \frac{x_i}{r}) e^{-(\mu/\rho)_i \rho x_i}   $$
********************************************************************
$$  \\   N_1 = N_0\frac{\Delta z_i}{nh} (1 + \frac{x_1}{r}) e^{-(\mu/\rho)_1 \rho x_i}$$
for second layer:
$$ \\   N_2 = N_1\frac{\Delta z_i}{nh} (1 + \frac{x_2}{r+x_1}) e^{-(\mu/\rho)_2 \rho x_2}$$
for third layer:
$$ \\   N_3 = N_2\frac{\Delta z_i}{nh} (1 + \frac{x_3}{r+x_1+x_2}) e^{-(\mu/\rho)_3 \rho x_3}$$

******************************************************
and if we substitute each value to the formulas like taht way:
$$ \\   N_2 = N_0\frac{(\Delta z_i)^2}{(nh)^2} (1+\frac{x_1}{r})(1 + \frac{x_2}{r+x_1}) e^{-(\mu/\rho)_1 \rho x_1-(\mu/\rho)_2 \rho x_2}$$
then 
$$ \\   N_3 = N_0\frac{(\Delta z_i)^3}{(nh)^3} (1+\frac{x_1}{r})(1 + \frac{x_2}{r+x_1})(1 + \frac{x_3}{r+x_1+x_2}) e^{-(\mu/\rho)_1 \rho x_1-(\mu/\rho)_2 \rho x_2 -(\mu/\rho)_3 \rho x_3}$$
******************************************************************************************
$$ \\   N_{n} = N_0\frac{(\Delta z_i)^{n-1}}{(nh)^{n-1}} \prod_i(1 + \frac{x_i}{r+\sum_{i}^{n} x_{i-1}})  e^{- \sum_{i}^{n}(\mu/\rho)_i \rho x_i}$$

```python
from functools import reduce
#start the calculation of Number of pacticles of photons source
P = 64              # Watt [W] source power
# P_254 = 20 # W Watt
# P_185 = 5 nm #Watt
hc = 1240          # eV
wave_length = 185    # wave length 185 nm
t = np.array([1,5,10,20,40,60,80,100,120])  # time in seconds
g_i=1 # for monochromatic source.
# 1J  = 0.624e-19 eV
nu=0.45 # % from electric power P
nu_184=0.30 # from P_uv =0.4*P_el
nu_254=0.75

#------------------------------------------------------------  
#input data
t = np.array([1,5,10,20,40,60,80,100,120])  # time in seconds
nu_lambda_arr = np.array([nu_184, nu_254])
wave_arr = np.array([wave_length, wave_length254])
#power_arr = np.array([P_uv184,P_uv254])
#--------------------class for Lamp-------------------------------------------------------------------
class PowerOfLamp:
    hc = 1240 #eV static parameter
    
    def __init__(self,power_el, nu_el, wave_length_arr, nu_lambda_arr):
        self.power_el = power_el
        self.nu_el = nu_el
        self.wave_length_arr = wave_length_arr
        self.nu_lambda_arr = nu_lambda_arr
        
    # method for calculation UV output power for different wave lengthes    
    def sum_of_power_arr(self):
        power_light = self.power_el*self.nu_el
        # using reduce method for multiplying power to each nu-coeficient and  addind them
        result = reduce(lambda a,b: a*power_light + b*power_light,self.nu_lambda_arr)
        return result
    
    # return array of efficincy for power at different wave lengthes    
    def power_arr_by_lambda(self):
        arr=np.array([]) # empty array for accumulation 
        for i in self.nu_lambda_arr:
            s = i*self.power_el*self.nu_el
            arr = np.append(arr,s)
        return arr
    # this method is static, and you can use it for summation of photons(photons flux)  for different wavelengthes and         
    @staticmethod   
    def sum_of_photons(power_arr, wave_arr):
        total = 0
        length = len(power_arr)
        for i in range(length):
            a=((power_arr[i]*wave_arr[i])/hc)*0.624e19
            print(a)
            total += a
            
        return total
    # return fluence of photon per each time interval t, return array of phton's fluences
    def number_photons(self,total_power, t):
        arr = np.array([])
        for i in t:
            n = total_power*i
            arr = np.append(arr,n)
        return arr

#---------------------------------------------------------------------------------------
lamp_64W = PowerOfLamp(P, nu, wave_arr,nu_lambda_arr)
#lamp_64W.sum_of_power_arr()     # 30.24 W for electrical_power = 64W
power_arr = lamp_64W.power_arr_by_lambda()
photons_in_1sec = lamp_64W.sum_of_photons(power_arr,wave_arr)     #3.565254193548387e+19 #(flux (s^-1)

print('Phtons per second:',photons_in_1sec)
print('--------with time exposure--------------')
number_of_photon_arr = lamp_64W.number_photons(photons_in_1sec,t)
print('Array of photons from 1s to 120s:\n ',number_of_photon_arr)
```

## Mass attenuation coefficient
$$ N_{am}=\frac{N_a}{A_r} $$ 
* Number  $N_{am}$ - number of atoms per unit mass of elemnt, $N_{a}$ - Avagadro Number (per mole or Ar grams), $A_{r}$- relative atomic number.
**************************************
$$ N_{e\rho}=\rho Z N_{am}=\rho Z \frac{N_a}{A_r} $$ 
*****************************
* Number density of of molecules in liquid water (we use this term for geometrical size of nucleus that within molecule) 
$$ n=\rho \frac{N_a}{\sum_{i}^{k}A_{ri}} $$ 
**************************************
$$ \mu = \sigma_{at}n + \sigma_e N_{e\rho}$$
* where $\sigma_{at}$ - aproximate size of nuclues of different atoms of average size ($\sigma_{at}
=\frac{\sigma_1+\sigma_2 \cdots+\sigma_N }{N}$ ) and $\sigma_e = \sigma_{ph+ion}+\sigma_{scat}$
* $\mu$ - linear attenuation coefficient
* $\sigma_{ph+ion}$ - cross section of photoeffect and ionistion of molecule (atom. $\sigma_{scat} $-scattering of photons per electon.
***************************
Or if we use mass attenuation coefficient ($\frac{\mu}{\rho}$) we can use:
* Number of electrons per unit mass
$$ N_{e}= Z \frac{N_a}{A_r} $$ 
********************************
* Number density per unit masss (number of atoms per unit mass)
$$ n_{\rho}= \frac{N_a}{\sum_{i}^{k}A_{ri}} $$ 
**************************************
* mass-attenuation coefficient
$$ (\frac{\mu}{\rho})_i= \sigma_{at}n_{\rho} + \sigma_e N_{e}$$

```python
# Number of electrons N_e
molec_of_water = ['H', 'O']
molec_of_indexi = [2, 1]
# Atomic number=Charge
Z_H=1
Z_O=8
# relative atomic masses
Ar_H = 1
Ar_O = 16
arr_ArH2O = np.array([Ar_H, Ar_O ])
arr_Z=np.array([Z_H, Z_O])
N_avg=6.0221408e+23 # Avogadro number
#---------------------------------------------
def calc_Ne(arr_Z, arr_Ar, arr_index):
    N_elm=np.array([]) # empty array of elements
    for i in range(0,len(arr_Z)):
        N_e=(arr_index[i]*arr_Z[i]*N_avg)/arr_Ar[i]
        N_elm = np.append(N_elm, N_e)
        print(N_elm)
    return np.sum(N_elm)

N_eO=calc_Ne(arr_Z, arr_ArH2O, molec_of_indexi)
print(N_eO)
```

## Ultraviolet Lamp Output Measurement: A Concise Derivation of the Keitz Equation

* This method relies on an equation, presented by Keitz (1971), relating the measured irradiance at a point to the total flux from the lamp. 
* Although this equation was presented by Keitz, the equation was not derived concisely in one section of the book; the necessary steps are scattered throughout various pages in text and figures (pp. 109, 136–138, 140–141). There is, therefore, a need for a clear derivation of the equation.
* The derivation models the lamp as a series of Lambertian, or cosine, emitters. Numerous investigators (Akehata and Shirai 1972; Bolton 2000; Ducoste et al. 2005; Liu et al. 2004) **have used forms of this model, but have differed in the *magnitude of the normal component of the emitted* radiation**. 
* Most have simply multiplied the irradiance *given by a point-source model by the cosine function* (e.g. equation 12 in Ducoste et al. 2005), but this is not correct since the cosine function is always less than or equal to unity, 
* and so the **total emitted power (which is equal to the irradiance integrated over an enclosing surface)** of the point cosine emitter would be less than that for a uniform point source. 
* Specifically, this failure to calculate the correct “normalization factor” results in calculated irradiance at any position that is a factor of **π/4** times its correct value. The present document will derive the correct **normalization factor**, and then derive a formula for determining the total flux from a UV lamp based on a single irradiance measurement. The resulting equation is analogous to the Keitz equation.
*************************
* Lambertian Emitters
* We will model the lamp as a **line source** with each differential element having **a cosine emission pattern** with respect to the angle from the lamp normal, θ, and being symmetric about the lamp axis. The emission pattern from a differential element is represented in Figure 1.
The flux per unit area, at a given distance from the midpoint of the differential element, can be expressed:
$$$$
$$
I = I_0 cos(\theta)
$$
$$$$
* where $I_0$ is the flux per unit area in the normal direction. We now will determine the value of this flux for the given power, $dP$, of the differential emitter element.
* Consider a spherical surface of radius r, centered on the emitter, that encloses the emitter. The divergence theorem states that the magnitude of the enclosed source (emitter) can be found by integrating the vector field (radiation) over any closed surface containing the source. If we had chosen a dif- ferent enclosing volume, we would have needed to consider the angle at which the radiation reaches the surface. However, because the sphere is centered on the element, all radiation reaches the surface normal to the surface. As a result, we can integrate the flux over the area without considering direction cosines. The geometry is shown in Figure 2.

sigma_scatC <- function(r_e, E_e, E_ph){
  alpha <- (E_ph/E_e)
  sigmaC <- ((r_e^2)/2)*(pi*(2*alpha^4+18*alpha^3+16*alpha^2+4*alpha*(4*alpha^4-4*alpha^3-15*alpha^2-10*alpha-2)*log(2*alpha+1))/(alpha^3)*(4*alpha^2+4*alpha+1))
  return (abs(sigmaC))
}
